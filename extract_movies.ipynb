```python
# Web Scrapping the Movie Datas

for i in range(0,n):
  sample =movie_urls[i]
  url = sample
  key_value_pairs = {}
  result = requests.get(url)

  if result.status_code == 200:
    soup = BeautifulSoup(result.content, 'lxml')
  else:
    print("Failed to fetch webpage. Status code:", result.status_code)
    continue

  temp = soup.find('span', class_='item mr-2')
  pair = temp.text.strip().split(':')
  key = pair[0].strip()
  value = pair[1].strip()
  key_value_pairs[key]=value


# Extracting Move Title
  heading = soup.find('h2', class_='heading-name')
  movie_title = heading.a.text.strip()
  key_value_pairs["MOVIE TITLE"]=movie_title

# Extract Release Date
  release_date_element = soup.find('span', class_='type', text='Released: ').find_next_sibling(text=True).strip()
  key_value_pairs['Release Date'] = release_date_element

# The Try-except command is used for dealing with missing data
# Extract Genre
  try:
    genre_element = soup.find('span', class_='type', text='Genre: ').find_next_sibling('a').text.strip()
    key_value_pairs['Genre'] = genre_element
  except Exception as e:
    key_value_pairs['Genre'] = "NA"

# Extract Cast
  try:
    cast_elements = soup.find('span', class_='type', text='Casts: ').find_next_siblings('a')
    cast_list = [element.text.strip() for element in cast_elements]
    key_value_pairs['Cast'] = cast_list
  except Exception as e:
    key_value_pairs['Cast'] = "NA"

# Extract Duration
  try:
    duration_element = soup.find('span', class_='type', text='Duration: ').find_next_sibling(text=True).strip()
    key_value_pairs['Duration'] = duration_element
  except Exception as e:
    key_value_pairs['Duration'] = "NA"

# Extract Country
  try:
    country_element = soup.find('span', class_='type', text='Country: ').find_next_sibling('a').text.strip()
    key_value_pairs['Country'] = country_element
  except Exception as e:
    key_value_pairs['Country'] = "NA"

# Extract Production company
  try:
    production_element = soup.find('span', class_='type', text='Production: ').find_next_sibling('a').text.strip()
    key_value_pairs['Production'] = production_element
  except Exception as e:
    key_value_pairs['Production'] = "NA"

  data.append(key_value_pairs)
```

```python
def scrape_movie_details(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        title = soup.find('h1', class_='title').text.strip()
        genre = soup.find('span', class_='genre').text.strip()
        rating = soup.find('div', class_='rating').text.strip()
        summary = soup.find('p', class_='summary').text.strip()

        # Store the movie details in a dictionary
        movie_details = {
            'Title': title,
            'Genre': genre,
            'Rating': rating,
            'Summary': summary,
            'URL': url
        }
        return movie_details
    
    else:
        print("Failed to fetch webpage. Status code:", response.status_code)
        return None
```

```python
# Extract Heading

for i in range (n):
  heading = soup.find('h2', class_='heading-name')
  movie_title = heading.a.text.strip()
  key_value_pairs["MOVIE TITLE"]=movie_title

  data.append(key_value_pairs)


# Extract Release Date
for i in range (n):
  release_date_element = soup.find('span', class_='type', text='Released: ')
  if release_date_element:
      release_date_value = release_date_element.find_next_sibling(text=True)
      if release_date_value:
          release_date_text = release_date_value.strip()
          key_value_pairs['Release Date'] = release_date_text
      else:
          key_value_pairs['Release Date'] = "Not Defined"

  data.append(key_value_pairs)
```

```python
# Extract cast
for i in range (n):
  cast_elements= soup.find('span', class_='type', text='Casts: ')
  if cast_elements:
      cast_anchors = cast_elements.find_next_siblings('a')
      if cast_anchors:
          cast_list = [element.text.strip() for element in cast_anchors]
          key_value_pairs['Cast'] = cast_list
      else:
          key_value_pairs['Cast'] = []

  data.append(key_value_pairs)
```

```python
# Extract duration
for i in range (n):
  duration_element = soup.find('span', class_='type', text='Duration: ')
  if duration_element:
      duration_value = duration_element.find_next_sibling(text=True)
      if duration_value:
          duration_text = duration_value.strip()
          key_value_pairs['Duration'] = duration_text
      else:
          key_value_pairs['Duration'] = "Not Defined"

  data.append(key_value_pairs)
```

```python
# Extract country
for i in range (n):
  country_element = soup.find('span', class_='type', text='Country: ')
  if country_element:
    country_value = country_element.find_next_sibling('a')
    if country_value:
        country_text = country_value.text.strip()
        key_value_pairs['Country'] = country_text
    else:
        key_value_pairs['Country'] = "No Country"

  data.append(key_value_pairs)
```

```python
# Extract production company
try:
    production_element = soup.find('span', class_='type', text='Production: ')
    if production_element:
        production_value = production_element.find_next_sibling('a')
        if production_value:
            production_text = production_value.text.strip()
            key_value_pairs['Production'] = production_text
        else:
            key_value_pairs['Production'] = "Not Defined"
    else:
        key_value_pairs['Production'] = "Not Defined"
    data.append(key_value_pairs)

except Exception as e:
    print("An error occurred while extracting production company:", e)
```
